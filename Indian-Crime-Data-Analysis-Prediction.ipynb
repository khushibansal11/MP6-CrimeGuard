# Kurtosis

Kurtosis is a measure of the "peakedness" or "flatness" of a probability distribution, which describes how much of the distribution is concentrated around the central point, and how much is spread out towards the tails.

A distribution with high kurtosis has a very sharp peak and heavy tails, indicating that it has more outliers and is more prone to extreme values. A distribution with low kurtosis is more flat-topped and has lighter tails, indicating that it has fewer outliers and is more stable.

# Time series analysis

```python
import statsmodels.api as sm
plt.figure(figsize=(15,8))

crime_df = pd.read_csv(r"crime-dataset.csv", parse_dates=['YEAR'], index_col=['YEAR'])

# Resample data to yearly frequency
crime_yearly = crime_df.resample('Y').sum()

# Plot the time series
plt.plot(crime_yearly.index, crime_yearly['TOTAL IPC CRIMES'])
plt.title('Total IPC Crimes over Time')
plt.xlabel('Year')
plt.ylabel('Number of Cases')
plt.show()

# Decompose the time series into trend, seasonality, and residuals
decomposition = sm.tsa.seasonal_decompose(crime_yearly['TOTAL IPC CRIMES'], model='additive')

# Plot the decomposition results
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(12, 7))
ax1.plot(crime_yearly.index, decomposition.observed)
ax1.set_ylabel('Observed')
ax2.plot(crime_yearly.index, decomposition.trend)
ax2.set_ylabel('Trend')
ax3.plot(crime_yearly.index, decomposition.seasonal)
ax3.set_ylabel('Seasonal')
ax4.plot(crime_yearly.index, decomposition.resid)
ax4.set_ylabel('Residual')
plt.tight_layout()
plt.show()

# Test for stationarity using the Augmented Dickey-Fuller test
adf_result = sm.tsa.stattools.adfuller(crime_yearly['TOTAL IPC CRIMES'])
print('ADF Statistic:', adf_result[0])
print('p-value:', adf_result[1])
print('Critical Values:', adf_result[4])
```

# Deep Analysis - Functions

### Query Analysis: Crime Overview by Year and State

```python
import pandas as pd

def query_analysis(year, state):
    # Filter dataset by year and state
    crime_filtered = crime[(crime['YEAR'] == year) & (crime['STATE/UT'] == state)]
    
    # Select the top 4 crimes (excluding total columns)
    crime_columns = ['MURDER', 'ATTEMPT TO MURDER', 'CULPABLE HOMICIDE NOT AMOUNTING TO MURDER',
                     'RAPE', 'KIDNAPPING & ABDUCTION', 'THEFT', 'ROBBERY', 'DOWRY DEATHS',
                     'CRUELTY BY HUSBAND OR HIS RELATIVES', 'ASSAULT ON WOMEN WITH INTENT TO OUTRAGE HER MODESTY']
    
    # Sum the crimes for the state and year
    crime_sum = crime_filtered[crime_columns].sum().sort_values(ascending=False)
    
    # Calculate percentages for the top 4 crimes
    total_crimes = crime_filtered['TOTAL IPC CRIMES'].sum()
    top_crimes = crime_sum.head(4)
    top_crimes_percentage = (top_crimes / total_crimes) * 100
    
    # Create a report
    report = pd.DataFrame({
        'Crime': top_crimes.index,
        'Number of Cases': top_crimes.values,
        'Percentage': top_crimes_percentage.values
    })
    
    print(f"Crime Analysis for {state} in {year}")
    # print(report)
    
    return report

# Example Usage
query_analysis(2005, 'MAHARASHTRA')
```

### Graph Analysis: Visualize Crime Data by Year and Type

```python
import matplotlib.pyplot as plt
import seaborn as sns

def graph_analysis(year, crime_type):
    # Filter data for the given year
    crime_filtered = crime[crime['YEAR'] == year]
    
    # Plot the crime type across states
    plt.figure(figsize=(10, 6))
    sns.barplot(x='STATE/UT', y=crime_type, data=crime_filtered, palette="viridis")
    plt.title(f"{crime_type} in different States for the year {year}")
    plt.xticks(rotation=90)
    plt.ylabel('Number of Cases')
    plt.xlabel('State')
    plt.tight_layout()
    plt.show()

# Example Usage
graph_analysis(2013, 'MURDER')
```

### Predictive Analysis: Forecast Future Crime Trends

```python
import statsmodels.api as sm

def predictive_analysis(state, year):
    # Filter data by state
    crime_state = crime[crime['STATE/UT'] == state]
    
    # Aggregate the data yearly for total IPC crimes
    crime_state_yearly = crime_state.groupby('YEAR').sum()['TOTAL IPC CRIMES']
    
    # Fit ARIMA model with best order (1,0,1)
    model = sm.tsa.ARIMA(crime_state_yearly, order=(1, 0, 1))
    model_fit = model.fit()
    
    # Forecast future crime numbers
    forecast = model_fit.forecast(steps=5)  # Predict next 5 years
    print(f"Future crime predictions for {state} starting from {year}:")
    print(forecast)

    # Plot predictions
    plt.figure(figsize=(10, 5))
    plt.plot(crime_state_yearly, marker='o', linestyle='-', label="Historical Data")
    plt.plot(range(crime_state_yearly.index[-1]+1, crime_state_yearly.index[-1]+6), forecast, marker='o', linestyle='--', color="red", label="Forecast")
    plt.title(f"Crime Forecast in {state}")
    plt.xlabel("Year")
    plt.ylabel("Total IPC Crimes")
    plt.legend()
    plt.grid(True)
    plt.show()
    
    return forecast

# Run with best ARIMA order
predictive_analysis('MAHARASHTRA', 2013)
```

### Conclusion and Reporting with Graphs

```python
import matplotlib.pyplot as plt

def conclusion_report(year, state):
    # Get data for the state and year
    report = query_analysis(year, state)
    
    # Plot the percentage contribution of the top 4 crimes
    plt.figure(figsize=(8, 6))
    plt.pie(report['Percentage'], labels=report['Crime'], autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
    plt.title(f"Crime Breakdown for {state} in {year}")
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()
    
    # Perform predictive analysis
    future_forecast = predictive_analysis(state, year)
    
    # Plot future predictions
    plt.figure(figsize=(10, 6))
    plt.plot(future_forecast, marker='o', linestyle='--', color='b')
    plt.title(f"Predicted Future Crimes for {state} (next 5 years)")
    plt.ylabel("Total IPC Crimes")
    plt.xlabel("Years")
    plt.grid(True)
    plt.show()

# Example Usage
conclusion_report(2013, 'MAHARASHTRA')
```

```python
min_val = crime['TOTAL IPC CRIMES'].min()
max_val = crime['TOTAL IPC CRIMES'].max()
range_val = (max_val - min_val) / 4
low = min_val + range_val
medium = low + range_val
high = medium + range_val
```

```python
def get_crime_level(crime_count):
    if crime_count <= low:
        return 1
    elif crime_count <= medium:
        return 2
    elif crime_count <= high:
        return 3
    else:
        return 4

crime['CRIME_LEVEL'] = crime['TOTAL IPC CRIMES'].apply(get_crime_level)
```

# Linear Regression

```python
X = crime.drop(['CRIME_LEVEL','STATE/UT', 'DISTRICT','TOTAL IPC CRIMES'], axis=1)
y = crime['TOTAL IPC CRIMES']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
lr = LinearRegression()
lr.fit(X_train,y_train)
lr_pred = lr.predict(X_test)
lr_score = lr.score(X_test, y_test)
print('Linear Regression score : ',lr_score)
print()
print('Mean_absolute_error  = ',metrics.mean_absolute_error(lr_pred,y_test))
print()
print('Mean_squared_error   = ',metrics.mean_squared_error(lr_pred,y_test))
print()
print('R2_score             = ',metrics.r2_score(lr_pred,y_test))
print()
R2 = metrics.r2_score(lr_pred,y_test)
adj_R2 = 1-((1-R2)*(len(y)-1)/(len(y)-X.shape[1]-1))
print('Adjusted_R2         = ',adj_R2)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```
